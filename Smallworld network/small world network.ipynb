### Produce WS network and measure chariacteristic path length & clustering coefficient as a fn(t) of p.

```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
import time
```

```python
def generate_network_ring(N):
    J = np.zeros((N,N))
    k = 0
    for a in range(N):
        i = a%N
        j = (a+1)%N 
        J[i,j] = 1
        J[j,i] = 1
    return J

def generate_network_k_ring(N,n):
    J = np.zeros((N,N))
    k = 0
    for a in range(N):
        i = a%N
        for k in range(n):
            j = (a+1+k)%N 
            J[i,j] = 1
            J[j,i] = 1
    return J

def generate_network_ER(N,p):
    J = np.zeros((N,N))
    k = 0
    for i in range(N):
        for j in range(i):
            if np.random.rand() < p:
                J[i,j] = 1
                J[j,i] = 1
    return J


def make_shortcut(A):
    B = np.array(A)
    while np.all(A == B):
        i,j = np.random.choice(N,2,replace = False)
        if A[i,j] == 0:
            A[i,j] +=1
            A[j,i] +=1
    return A



def rewiring(A):
    N = len(A)
    B = np.array(A)
    
    while np.all(A == B):
        ## 이 process 가 네트워크에 변화를 주면 process 를 끝냄.
        i,j = np.random.choice(N,2,replace = False)
        
        if A[i,j] == 0 and np.any(A[i,:]==1):
            ## i,j 를 선택하고 이 두 노드가 연결이 안되어있다면, 이 둘을 연결하고, i 와 연결된 임의의 노드를 끊어줌.
            a = np.random.choice(np.where(A[i,:]==1)[0],size = 1)
            A[i,a] = 0
            A[a,i] = 0
            A[i,j] =1
            A[j,i] =1 
            
    return A


    
def clust_coeff(A):
    N = len(A)
    Num_of_triple = 0.
    Num_of_clos_triple = 0.
    for i in range(N):
        B = np.where(A[i,:]==1)[0]
        
        for j in range(len(B)):
            for k in range(j):
                if A[B[j],B[k]] == 1:
                    Num_of_clos_triple+=1
                Num_of_triple += 2
    Num_of_clos_triple *= 2
    if Num_of_triple == 0:
        c = 0.
    else:
        c = Num_of_clos_triple/Num_of_triple
    return c


def clust_coeff_mat(A):
    Num_of_clos_triple = np.trace(A@A@A)
    degree = np.sum(A, axis = 1)
    if np.sum(degree*(degree-1)) ==0:
        c = 0
    else:
        c = Num_of_clos_triple/np.sum(degree*(degree-1))
    return c

def shortest_path(A):
    N = len(A)
    B = np.array(A) ## B_i,j = i,j th node shortest path length
    C = np.array(A)
    for i in range(N-1):
        Q = np.zeros((N,N))
        C = C@A
        B += np.where(np.where(C!=0,1.,0)+np.where(B==0,1.,0)>1.5,(i+2), 0)
        if 0 not in B:
            break
    return B
```

### ER network 

```python
N = 70
p = 0.05
A = generate_network_ER(N,p)
G = nx.Graph(A)
pos = nx.circular_layout(G)
nx.draw_networkx(G,pos=pos,font_size  = 20,node_color ='y')
ax = plt.gca()
ax.set_aspect('equal')
plt.axis('off')
plt.show()

c = clust_coeff_mat(A)
print('Clustering Coefficent : ',c)

iu1 = np.triu_indices(N,1)
B = shortest_path(A)
aL = np.average(B[iu1])
print('average_shortest_path_length',aL)

l = nx.average_shortest_path_length(G)
print('average_shortest_path_length',l)
```
